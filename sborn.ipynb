```python
# Spyder Installation
#   https://pythonhosted.org/spyder/installation.html
# Seaborn Overview
#   https://seaborn.pydata.org/index.html
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
import pandas as pd
%matplotlib inline  
```

```python
#==============================================================================
# Seaborn Settings
#==============================================================================

# Options are "paper", "talk" and "poster"
sns.set_context('talk')
# Options are "white", "whitegrid", "dark", "darkgrid" and "ticks"
sns.set_style('ticks')

# Huge list of colors at
# https://xkcd.com/color/rgb/
color = sns.xkcd_rgb['vomit green']
```

```python
#==============================================================================
# Generate a random correlated bivariate dataset (2D Gaussian)
#==============================================================================

# Methods for generating random numbers
rs = np.random.RandomState(42)
# Parameters needed to generate the 2D Gaussian
mean = [0, 0]
cov = [(1, .7), (.7, 1)]
x1, x2 = rs.multivariate_normal(mean, cov, 1000).T

# This is just because seaborn can read axis names from panda object
x1 = pd.Series(x1, name='Who Knows $\Omega$')
x2 = pd.Series(x2, name='Whatever $\phi \pi$')
```

```python
#==============================================================================
# Plot generated points
#==============================================================================

f1 = plt.figure(figsize=(7,7))
#plt.scatter(x1, x2, marker='o', s=15, color=color)

# With colorbar based on a third array
x3 = np.random.normal(loc=0, scale=5, size=len(x1))
plt.scatter(x1, x2, marker='o', s=15, c=x3, cmap=cm.terrain)

plt.colorbar()
plt.xlabel(x1.name)
plt.ylabel(x2.name)
f1.tight_layout()
plt.show()
```

```python
#==============================================================================
# A marginal distribution
#==============================================================================

f2 = plt.figure(figsize=(7,7))
sns.distplot(x1, rug=True, hist=True, kde=True, norm_hist=True)
f2.tight_layout()
```

```python
#==============================================================================
# Joint and marginal distribtuions in one
#==============================================================================

f3 = sns.jointplot(x1, x2, kind="kde", size=7, space=0)
```
